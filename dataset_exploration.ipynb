```python
import pandas as pd
import configparser
import json
config = configparser.ConfigParser()
config.read('config.ini')
folder_path = config["dataset"]["folder_path"]

df = pd.read_csv(f"{folder_path}\\raw\\twitch_target.csv", encoding='utf-8')
f = open(f"{folder_path}\\raw\\twitch_edges.json")
edges = json.load(f)
```

```python
df.head()
```

```python
def get_samples(num, class_num):
    sampled_class = df[df['target'] == class_num].sample(num)['id']
    sampled_edges = []
    for sample in sampled_class:
        sampled_edges.append(edges[str(sample)])

    return sampled_edges
```

```python
edges0 = get_samples(num = 5, class_num= 0)
edges1 = get_samples(num = 5, class_num= 1)
```

```python
import networkx as nx
G = nx.Graph()
G.add_edges_from(edges0[0])
nx.draw(G)
G.clear()
```

```python
G2 = nx.Graph()
G2.add_edges_from(edges1[0])
nx.draw(G2)
```
